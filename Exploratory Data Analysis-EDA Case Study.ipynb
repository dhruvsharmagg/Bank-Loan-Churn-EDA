## Dealing with missing data

```python
## Import libraries for plotting data
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Display boxplot for annuity amount 
sns.boxplot(x=df['AMT_ANNUITY'])
plt.show()
```

###### From the boxplot above, we can see that there are outliers present in the column AMT_ANNUITY. To deal with missing values, we need to replace them with the MEDIAN of the column

```python
## Replacing the NaN values with median for the column AMT_ANNUITY
df.AMT_ANNUITY.fillna(df.AMT_ANNUITY.median(),inplace=True)
```

```python
## Display boxplot for amount of goods price 
plt.figure(figsize=(8,7))
sns.boxplot(x=df['AMT_GOODS_PRICE'])
plt.show()
```

###### From the boxplot above, we can see that there are outliers present in the column AMT_GOODS_PRICE. To deal with missing values, we need to replace them with the MEDIAN of the column

```python
## Replacing the NaN values with median for the column AMT_GOODS_PRICE
df.AMT_GOODS_PRICE.fillna(df.AMT_GOODS_PRICE.median(),inplace=True)
```

```python
## Display the unique values in NAME_TYPE_SUITE
df.NAME_TYPE_SUITE.unique()
```

```python
## Display the most frequent value in the NAME_TYPE_SUITE
df.NAME_TYPE_SUITE.mode()[0]
```

###### Since NAME_TYPE_SUITE is a categorical column, the best approach which we have chosen here is to replace the missing values with the MODE of the column

```python
## Replace the NaN values with the most frequent value i.e. 'Unaccompanied'
df.NAME_TYPE_SUITE = df.NAME_TYPE_SUITE.fillna(df.NAME_TYPE_SUITE.mode()[0])
```

```python
## Display boxplot for amount of credits per week
sns.boxplot(x=df['AMT_REQ_CREDIT_BUREAU_WEEK'])
plt.show()
```

```python
## Display boxplot for count of family members
sns.boxplot(x=df['CNT_FAM_MEMBERS'])
plt.show()
```

###### From the boxplot above, we can see that there are no outliers present in the column AMT_REQ_CREDIT_BUREAU_WEEK  . To deal with missing values, we need to replace them with the MEAN of the column

```python
## Replacing the NaN values with mean for  the columns AMT_REQ_CREDIT_WEEK and CNT_FAM_MEMBERS
df.AMT_REQ_CREDIT_BUREAU_WEEK.fillna(df.AMT_REQ_CREDIT_BUREAU_WEEK.mean(),inplace=True)
df.CNT_FAM_MEMBERS.fillna(df.CNT_FAM_MEMBERS.mean(),inplace=True)
```

```python
## Visualize for the proportion of client with payment difficulties
labels = 'Customer with difficulty', 'Customer with no difficulty'
sizes = [df.TARGET[df['TARGET']==1].count(), df.TARGET[df['TARGET']==0].count()]
explode = (0, 0.1)
fig1, ax1 = plt.subplots(figsize=(10, 8))
ax1.pie(sizes,explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=110)
ax1.axis('equal')
plt.title("Proportion of client with payment difficulties", size = 20)
plt.show()
```

```python
print('Imbalance ratio in TARGET variable is {0}'.format(round(imbalance[0]/imbalance[1],2)))
```
